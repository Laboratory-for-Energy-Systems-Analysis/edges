```python
import matplotlib.pyplot as plt
# method = ('ImpactWorld+ 2.1', 'Land transformation, biodiversity', 'midpoint')
methods = [('ImpactWorld+ 2.1', 'Water availability, human health', 'damage'),]
methods = [('ImpactWorld+ 2.1', 'Freshwater ecotoxicity', 'midpoint'),]
for method in methods:
    print(method)
    LCA = EdgeLCIA({act: 1}, method)
    LCA.lci()

    LCA.apply_strategies()

    LCA.evaluate_cfs()
    LCA.lcia()
    df = LCA.generate_cf_table()
```
