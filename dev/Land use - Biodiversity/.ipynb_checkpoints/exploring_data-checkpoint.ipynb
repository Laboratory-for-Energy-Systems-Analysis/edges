# Exploring the data from Scherer_2023

https://zenodo.org/records/10114493

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
```

```python
# Open the .tif file
with rasterio.open("data/PDF/land_use.tif") as src:
    # Basic metadata
    print("Width, Height:", src.width, src.height)
    print("Number of Bands:", src.count)
    print("Coordinate Reference System (CRS):", src.crs)
    print("Bounds:", src.bounds)
    print("Data Type:", src.dtypes)

    # Full metadata
    print("Metadata:", src.meta)
```

```python
with rasterio.open("data/PDF/land_use.tif") as src:
    # Read the first band
    band1 = src.read(1)
    
    # Mask nodata values
    nodata = src.nodata
    band1_masked = np.ma.masked_equal(band1, nodata)

    # Plot the band
    plt.imshow(band1_masked, cmap="gray")
    plt.colorbar()
    plt.title("Band 1 (Masked)")
    plt.show()
```

```python
# Open the file and inspect metadata
with rasterio.open("data/PDF/land_use.tif") as src:
    print("Metadata:", src.meta)
    print("Descriptions:", src.descriptions)  # Check if bands have descriptions
    print("Tags:", src.tags())  # Check for embedded tags
```

```python
with rasterio.open("data/PDF/land_use.tif") as src:
    # Read the first band
    band1 = src.read(1)
    
    # Check unique values
    unique_values = np.unique(band1)
    print("Unique values:", unique_values[:10])  # Display the first 10 unique values
    print("Number of unique values:", len(unique_values))
```

```python
land_use_labels = [
    "Primary Vegetation Minimal",
    "Cropland Intense",
    "Cropland Light",
    "Cropland Minimal",
    "Managed Forest Intense",
    "Managed Forest Light",
    "Managed Forest Minimal",
    "Pasture Intense",
    "Pasture Light",
    "Pasture Minimal",
    "Plantation Intense",
    "Plantation Light",
    "Plantation Minimal",
    "Urban Intense",
    "Urban Light",
    "Urban Minimal",
]

# Define colors for each category
cmap = mcolors.ListedColormap(['white', 'yellow', 'orange', 'tan', 'brown', 'darkgreen',
                                'lightgreen', 'darkblue', 'blue', 'lightblue', 'purple',
                                'pink', 'violet', 'red', 'darkred', 'gray'])
bounds = np.arange(1, 18)  # 1 to 16 land use codes, plus nodata
norm = mcolors.BoundaryNorm(bounds, cmap.N)

with rasterio.open("data/PDF/land_use.tif") as src:
    band1 = src.read(1)

    # Mask nodata values
    band1_masked = np.ma.masked_equal(band1, -3.4e+38)

    # Plot the data
    plt.figure(figsize=(10, 6))
    plt.imshow(band1_masked, cmap=cmap, norm=norm)
    cbar = plt.colorbar(ticks=np.arange(1, 17))
    cbar.ax.set_yticklabels(land_use_labels)
    plt.title("Land Use Types and Intensities")
    plt.show()
```

```python
# Count occurrences of each land use type
unique, counts = np.unique(band1[band1 != -3.4e+38], return_counts=True)
data_distribution = pd.DataFrame({'Value': unique, 'Count': counts})
data_distribution['Land Use Type'] = [land_use_labels[int(v) - 1] for v in unique]  # Map labels
print(data_distribution)
```

```python
with rasterio.open("data/PDF/land_use.tif") as src:
    # Check for embedded attributes or tags
    tags = src.tags()
    print("Tags:", tags)
```

# Mining Data - Jasansky

https://zenodo.org/records/7369478

### Exploring facilities.gpkg

```python
gpkg_path = "data/mining/jasansky/data/facilities.gpkg"  # Replace with the actual file path
facilities = gpd.read_file(gpkg_path)

# Inspect the data
print(facilities.head())
print("Columns:", facilities.columns)
print("CRS:", facilities.crs)
```

```python
facilities.plot(figsize=(10, 6), color="orange", edgecolor="black")
plt.title("Mining Facilities")
plt.show()
```

```python
print("Unique primary commodities:")
print(facilities['primary_commodity'].unique())

# Unique commodities/products
print("\nUnique commodities/products:")
print(facilities['commodities_products'].unique())
```

```python
# Count facility types
facility_types = facilities['facility_type'].value_counts()
print("\nFacility types and counts:")
print(facility_types)
```

```python
# Earliest and latest production start dates
print("\nProduction start dates:")
print(facilities['production_start'].min(), "to", facilities['production_start'].max())

# Count active facilities by year
active_status = facilities['activity_status'].value_counts()
print("\nFacility activity status:")
print(active_status)
```

```python
len(facilities)
```

```python
from tabulate import tabulate

# Step 1: Select 10 random mines
random_mines = facilities.sample(10, random_state=11)  # Set random_state for reproducibility

# Step 2: Plot their location
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the world basemap
world_path = "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(world_path)
world.plot(ax=ax, color='lightgray', edgecolor='black')

# Plot the random mines
random_mines.plot(ax=ax, color='red', markersize=50, label="Random Mines")

# Add plot title and legend
plt.title("Locations of 10 Random Mining Facilities", fontsize=14)
plt.legend()
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Step 3: Print a nicely formatted table
# Truncate the table to relevant columns
columns_to_display = [
    # "facility_name", 
    "facility_type", 
    "primary_commodity", 
    "commodities_products", 
    "country", 
    # "production_start", 
    # "production_end", 
    'activity_status', 
    'surface_area_sq_km'
]
truncated_table = random_mines[columns_to_display]

# Print the table using tabulate
print("\nTable of 10 Random Mining Facilities:")
print(tabulate(truncated_table, headers='keys', tablefmt='fancy_grid'))
```
