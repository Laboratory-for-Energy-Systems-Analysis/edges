```python
import pandas as pd
# open AWARE2018_total_country_cons_ecoinvent310.xlsx
# to get water consumption per country
water_use = pd.read_excel("AWARE2018_total_country_cons_ecoinvent310.xlsx", index_col=0)
```

```python
for category, cf in cfs.items():
    water_use_subset = water_use[category]
    # turn dataframe to dictionary, using index as keys
    water_use_dict = water_use_subset.to_dict()
    for c in cf:
        c["weight"] = water_use_dict.get(c["consumer"]["location"], 0)
```
